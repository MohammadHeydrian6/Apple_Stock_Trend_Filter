```python
import pandas as pd
import matplotlib.pyplot as plt


file_path = 'AAPL.xlsx'
df = pd.read_excel(file_path)

# هموارسازی با روش میانگین
smoothed_data_mean = df['Adj Close'].rolling(window=5).mean().dropna()

# هموارسازی با روش کران
smoothed_data_quantile = df['Adj Close'].rolling(window=5).quantile(0.5).dropna()


print("Smoothed Data (Mean):")
print(smoothed_data_mean.reshape(-1, 5))



print("\nSmoothed Data (Quantile):")
print(smoothed_data_quantile.values.reshape(-1, 5))


# رسم نمودارها
plt.figure(figsize=(10, 6))
plt.plot(df['Adj Close'], label='Raw Data')
plt.plot(smoothed_data_mean, label='Smoothed Data (Mean)')
plt.plot(smoothed_data_quantile, label='Smoothed Data (Quantile)')
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Comparison of Raw and Smoothed Data')
plt.legend()
plt.show()
```
